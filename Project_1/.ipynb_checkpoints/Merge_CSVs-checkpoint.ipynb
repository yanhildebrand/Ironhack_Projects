```python
import pandas as pd
import os

# List of CSV files to merge
csv_files = ['CAvideos.csv', 'DEvideos.csv', 'FRvideos.csv',\
             'USvideos.csv',\
             'GBvideos.csv',\
             'JPvideos_new.csv'\
            ]

# Create an empty DataFrame to store the merged data
merged_data = pd.DataFrame()

# Iterate over the CSV files
for file in csv_files:
    # Read each CSV file
    data = pd.read_csv(file)
    
    # Extract the file name
    file_name = os.path.basename(file)
    
    # Add the file name to a new column
    data['Country'] = file_name[:2]
    
    # Concatenate the data frames
    merged_data = pd.concat([merged_data, data], ignore_index=True)



# Write the merged data to a new CSV file

merged_data.to_csv('merged_test_6.csv', index=False)
```

```python
df = pd.read_csv('merged_test_6.csv')
df
```

```python
# replace names
df['Country'] = df['Country'].replace({'CA': 'Canada', 'US': 'United States','JP': 'Japan', 'DE': 'Germany', 'FR': 'France', 'GB': 'Great Britain'})
df
```

```python
# Delete columns
columns_to_delete = ['comments_disabled', 'ratings_disabled','video_error_or_removed']
df = df.drop(columns=columns_to_delete)
df
```

```python
df.to_csv('data_tableau.csv', index=False)
```
